Выполнил студент группы ПСАПР-25-2 Оберемок Артём

• Объявляем глобальные переменные

```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
```

• Пишем функцию для первой задачи

```python
def who_is_owner(number): # найти владельца документа
    result = 'Владелец документа: '
    for doc in documents:
        if doc['number'] == number:
            result += f'{doc['name']}'
            return result
    result += 'владелец не найден'
    return result
```

• Пишем функцию для второй задачи

```python
def where(number): # найти полку, на которой находится документ
    result = 'Документ хранится на полке: '
    for shelf in directories:
        if number in directories[shelf]:
            result += f'{shelf}'
            return result
    result += 'полка не найдена'
    return result
```

• Пишем основную часть программы

```python
while True:
    command = input('Введите команду: ')
    if command != 'q':
        number = input('Введите номер документа: ')
        if command == 'p':
            print(who_is_owner(number))
        if command == 's':
            print(where(number))
    else: break
```
