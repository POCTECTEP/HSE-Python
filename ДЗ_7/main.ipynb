Пишем класс и функции к нему

```python
"""    Класс для работы с данными пользователя    """
class User:
    def __init__(self, full_name, sex, age, device_type, browser, summ, region):
        self.full_name = full_name # ФИО
        self.sex = sex # Пол
        self.age = age # Возраст
        self.device = device_type # Устройство, с которого выполнялась покупка
        self.browser = browser # Браузер
        self.summ = summ # Сумма чека
        self.region = region # Регион покупки

    def make(self):
        result = 'совершил'
        if self.sex == 'female':
            result += 'а'
        return result

    def sex_result(self):
        if self.sex == 'female':
            return 'женского'
        return 'мужского'

    def age_result(self):
        if self.age % 10 == 1 and not 11 <= self.age%100 <= 14:
            return 'год'
        elif 2 <= self.age % 10 <= 4 and not 11 <= self.age%100 <= 14:
            return 'года'
        return 'лет'

    def region_result(self):
        if self.region == '-':
            return ' неопределён'
        return f': {self.region}'

    def device_result(self):
        if self.device in ["mobile", "tablet"]:
            return "мобильного"
        return "настольного"

    def description(self):
        return f"Пользователь {self.full_name} {self.sex_result()} пола, {self.age} {self.age_result()} {self.make()} покупку на {self.summ} у.е. с {self.device_result()} браузера {self.browser}. Регион, из которого совершалась покупка{self.region_result()}."
```

Проверяем пример из файла

```python
user = User("Allen Miss. Elisabeth Walton", "female", 29, "mobile", "Chrome", 885, "St Louis: MO")
print(user.description())
```

```python
print(user.description() == "Пользователь Allen Miss. Elisabeth Walton женского пола, 29 лет совершила покупку на 885 у.е. с мобильного браузера Chrome. Регион, из которого совершалась покупка: St Louis: MO.")
```

Ответ программы полностью совпадает с эталонным

Пишем основную часть кода

```python
data = open('web_clients_correct.csv')
data.readline()

with open("descriptions.txt", "w", encoding="utf-8") as f:
    for i in data:
        components = (i.replace('\n','').split(','))
        full_name = components[0]
        device_type = components[1]
        browser = components[2]
        sex = components[3]
        age = int(components[4])
        summ = int(components[5])
        region = components[6]
        f.write(f'{User(full_name, sex, age, device_type, browser, summ, region).description()}\n')
```
