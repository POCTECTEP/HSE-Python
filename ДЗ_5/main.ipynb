Код, принимающий дату

```python
from datetime import *

formats = {
    "Wednesday, October 2, 2002": "%A, %B %d, %Y",
    "Friday, 11.10.13": "%A, %d.%m.%y",
    "Thursday, 18 August 1977": "%A, %d %B %Y"
}

while True:
    try:
        date_str = input()
        date = datetime.strptime(date_str, formats[date_str])
        print(date)
    except: print("Формат неизвестен")
```
