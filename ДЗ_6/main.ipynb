и вручную расставим []

Код программы

```python
import json

visiters = {}
purchase_log = json.load(open('purchase2_log.json'))
visit_log = (open('visit_log__1_.csv'))
for i in visit_log:
    i = i.split(',')
    user_id = i[0]
    source = i[1]
    visiters[user_id] = source

with open('funnel.csv', "w", encoding="utf-8") as f:
    for i in purchase_log:
        user_id = i['user_id']
        category = i['category']
        try:
            result = f'{user_id},{str(visiters[user_id]).replace('\n','')},{category}\n'
            f.write(result)
        except:pass
```
